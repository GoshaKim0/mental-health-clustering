Step 1 - Load dataset and quick overview

```python
#
import pandas as pd
import numpy as np
from IPython.display import display

# Display config (readability only; does not change data)
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 120)

DATA_PATH = "mental-heath-i-tech-2016_20161114.csv"  # relative path

# Robust CSV read (fallback encoding if needed)
try:
    df = pd.read_csv(DATA_PATH)
except UnicodeDecodeError:
    df = pd.read_csv(DATA_PATH, encoding="latin1")

# Basic shape/columns preview
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("Columns (first 10):", list(df.columns[:10]))

# Peek first rows
display(df.head(10))

# Dtypes and top missing-value columns
display(df.dtypes.to_frame("dtype"))
missing = df.isna().sum().sort_values(ascending=False)
display(missing.head(20).to_frame("na_count"))
```

Step 2 - Split features by dtype (categorical vs numeric)

```python
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
numeric_cols = df.select_dtypes(exclude=["object"]).columns.tolist()

print(f"Categorical columns: {len(categorical_cols)}")
print(f"Numeric columns: {len(numeric_cols)}")

# Quick view of the first 10 categorical columns
display(df[categorical_cols[:10]].head())

# Check unique values count (helps detect free-text answers)
unique_counts = df[categorical_cols].nunique().sort_values(ascending=False)
display(unique_counts.head(20).to_frame("unique_count"))
```

Step 3 - Missing values overview (helps decide what to drop or impute)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort columns by % of missing
missing_percent = df.isnull().mean().sort_values(ascending=False) * 100
display(missing_percent.head(10).to_frame("missing_%"))

# Quick visualization
plt.figure(figsize=(10, 4))
sns.histplot(missing_percent, bins=20, kde=False)
plt.title("Distribution of missing values (%) across columns")
plt.xlabel("% missing")
plt.ylabel("Number of columns")
plt.show()
```

Step 4 - Drop columns with too many missing values (>70%)

```python
threshold = 0.7
cols_to_drop = missing_percent[missing_percent > threshold * 100].index.tolist()
print(f"Dropping {len(cols_to_drop)} columns with >70% missing values")

df = df.drop(columns=cols_to_drop)
print(f"New shape: {df.shape}")
```

Step 5 - Check remaining missing values after column drop

```python
remaining_missing = df.isnull().mean().sort_values(ascending=False) * 100
display(remaining_missing[remaining_missing > 0].head(10).to_frame("missing_%"))
```

Step 6 - Check how many columns still have missing values (sanity check)

```python
remaining_missing = df.isnull().mean() * 100
cols_with_missing = (remaining_missing > 0).sum()
print(f"Columns still having missing values: {cols_with_missing}")

plt.figure(figsize=(8,4))
sns.histplot(remaining_missing[remaining_missing > 0], bins=15, kde=False)
plt.title("Remaining missing values after drop (>0%)")
plt.xlabel("% missing")
plt.ylabel("Number of columns")
plt.show()
```

Step 7 - Simple imputation for remaining missing values

```python
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# Final check: should be 0 missing
print("Total missing after imputation:", df.isnull().sum().sum())
```

Step 8 - Basic numeric summary and outlier inspection

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary stats for numeric features
display(df.describe().T)

# Quick boxplots for each numeric column
for col in df.select_dtypes(exclude="object").columns:
    plt.figure(figsize=(6, 2))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot - {col}")
    plt.show()
```

Step 9 - Clean unrealistic ages

```python
age_before = df.shape[0]
df = df[(df["What is your age?"] >= 10) & (df["What is your age?"] <= 100)]
age_after = df.shape[0]

print(f"Removed {age_before - age_after} rows with unrealistic ages")
print(f"New shape: {df.shape}")
```

Step 10 - Encode categorical variables

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

# Simple label encoding for each categorical column
le = LabelEncoder()
for col in df_encoded.select_dtypes(include="object").columns:
    df_encoded[col] = le.fit_transform(df_encoded[col])

print(f"Encoded dataframe shape: {df_encoded.shape}")
display(df_encoded.head())
```

Step 11 - Feature scaling for PCA and clustering

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded)

print("Scaled matrix shape:", X_scaled.shape)
print("Mean:", round(X_scaled.mean(), 3), "| Std:", round(X_scaled.std(), 3))
```

Step 12 - PCA for dimensionality reduction and visualization

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Fit PCA
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_scaled)

# Explained variance plot
plt.figure(figsize=(8,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.title("PCA - Explained Variance Ratio")
plt.grid(True)
plt.show()

print("Variance explained by first 2 components:", round(np.sum(pca.explained_variance_ratio_[:2])*100, 2), "%")
```

Step 13 - Visualize PCA (2D projection)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=20, alpha=0.6)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("2D Projection of Respondents - PCA Space")
plt.grid(True)
plt.show()
```

Step 14 - Find optimal number of clusters using KMeans + silhouette

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

sil_scores = []
k_values = range(2, 10)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_pca[:, :5])  # use first 5 PCA comps for stability
    sil = silhouette_score(X_pca[:, :5], labels)
    sil_scores.append(sil)

plt.figure(figsize=(8,4))
plt.plot(k_values, sil_scores, marker='o')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette score")
plt.title("Silhouette Analysis for KMeans")
plt.grid(True)
plt.show()

best_k = k_values[np.argmax(sil_scores)]
print(f"Best k (highest silhouette): {best_k}")
```

Step 15 - Final KMeans clustering and visualization

```python
best_k = 4
kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
labels_final = kmeans_final.fit_predict(X_pca[:, :5])

# Add cluster labels to DataFrame
df["cluster"] = labels_final

# Cluster sizes
print(df["cluster"].value_counts().sort_index())

# Visualize clusters in PCA 2D space
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_final, cmap="viridis", s=25, alpha=0.7)
plt.title(f"KMeans Clusters (k={best_k}) in PCA space")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster")
plt.show()
```

Step 17 - K-Prototypes cluster distribution

```python
from kmodes.kprototypes import KPrototypes

categorical_cols = df.select_dtypes(include='object').columns
numerical_cols = df.select_dtypes(exclude='object').columns

X = df.copy()
for col in categorical_cols:
    X[col] = X[col].astype('category').cat.codes

X_matrix = X.to_numpy()
cat_idx = [X.columns.get_loc(c) for c in categorical_cols]

kproto = KPrototypes(n_clusters=4, init='Cao', random_state=42)
clusters_kproto = kproto.fit_predict(X_matrix, categorical=cat_idx)
df["cluster_kproto"] = clusters_kproto

print("K-Prototypes cluster distribution:")
print(df["cluster_kproto"].value_counts().sort_index())
```

Step 18 - t-SNE projection for K-Prototypes clusters (UMAP fallback)

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

# convert to float for TSNE
X_tsne = X_matrix.astype(np.float32)

tsne = TSNE(n_components=2, random_state=42, perplexity=30, learning_rate=200)
embedding = tsne.fit_transform(X_tsne)

plt.figure(figsize=(8, 6))
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=df["cluster_kproto"],
    cmap="tab10",
    s=25,
    alpha=0.8
)
plt.title("t-SNE projection - K-Prototypes clusters")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.grid(alpha=0.3)
plt.show()
```

Step 19 - Pairwise PCA plot

```python
import seaborn as sns
import pandas as pd

pca_df = pd.DataFrame(X_pca[:, :3], columns=["PC1", "PC2", "PC3"])
pca_df["cluster"] = labels_final

sns.pairplot(pca_df, hue="cluster", diag_kind="kde", palette="viridis")
plt.suptitle("Pairwise PCA Components by Cluster", y=1.02)
plt.show()
```
